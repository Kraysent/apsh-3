Более формально:
- У нас есть две точки, начальное положение и скорости которых на двумерной плоскости задано.
- В каждый момент времени их скорость по каждой оси меняется на величину, определяемую их ускорением по этой оси.
- Это ускорение определяется тем, как в этот момент на них действует ньютоновская сила гравитации.
- Так же, в каждый момент времени их положение по каждой оси меняется на величину, определяемую тем, какую скорость они имеют в этот момент.

В каждый момент времени мы можем считать, что изменения положения и скорости в следующий момент времени (который отстоит от текущего на очень маленькую величину) меняется как (для каждого тела):
- $v_{x (i+1)} = a_{x (i)} \Delta t$
- $v_{y (i+1)} = a_{y (i)} \Delta t$
- $x_{i+1} = v_{x (i)} \Delta t$
- $y_{i+1} = v_{y (i)} \Delta t$

А ускорение в любой момент времени высчитывается из сил гравитации:

$$
F_{x} = G \frac{m_1 m_2 (x_2 - x_1)}{r^3}
$$
$$
F_{y} = G \frac{m_1 m_2 (y_2 - y_1)}{r^3}
$$

где $r = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}$

$$
a_{1x} = F_x / m_1
$$
$$
a_{1y} = F_y / m_1
$$
$$
a_{2x} = F_x / m_2
$$
$$
a_{2y} = F_y / m_2
$$
