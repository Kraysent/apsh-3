#### Разница между гистограммой и столбчатой диаграммой

- Столбчатая диаграмма показывает, как распределено какое-то число в зависимости от *дискретного* значения. Дискретный, т.е. прерывистый - существует конечное число уникальных значений той величины, а зависимости от которой мы строим. В столбчатой диаграмме порядок значений на оси X чаще всего не важен.
- Гистограмма показывает распределение одной величины от *непрерывной* другой величины. Чтобы построить гистограмму, нужно перед построением разложить данные по "бакетам" (или "бинам") - дискретным диапазонам той величины, от которой мы строим зависимость. В гистограммах важно сохранить порядок величины, от которой идёт построение зависимости.

Для рисования столбчатых диаграмм в `matplotlib` есть отдельная функция (https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html):

```python
import matplotlib.pyplot as plt

classes = ["Ia", "Ibc", "II"]
percents = [25, 15, 60]

plt.bar(classes, percents)
plt.xlabel("Тип сверхновой")
plt.ylabel("Число событий, %")
plt.show()
```

Для гистограмм так же есть отдельная функция (https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html):

```python
import numpy as np
import matplotlib.pyplot as plt

# Гауссово (нормальное) распределение
data = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(data, bins=30) # bins - число бакетов/бинов, по которым разложить случайную величину
plt.xlabel("Значение случайной величины")
plt.ylabel("Число")
plt.show()
```

С гистограммами и столбчатыми диаграммами можно играться:

```python
import numpy as np
import matplotlib.pyplot as plt

# Случайная величина с экспоненциальным распределенем - вероятность выпадения величины x спадает по экспоненте
data = np.random.exponential(scale=10, size=1000)

plt.hist(data, bins=30, orientation='horizontal', edgecolor='red', facecolor='white')
plt.xlabel("Число")
plt.ylabel("Значение случайной величины")
plt.show()
```

### Фильтрация данных в `numpy`

<div class="alert alert-block alert-warning" style="margin-top: 20px">

<font size=4>**Задание 1**</font>

По данным спектральных классов построить гистограмму числа звёзд в каждом классе. Использовать логарифмическую шкалу.

</div>

```python
import numpy as np
import matplotlib.pyplot as plt

arr = np.genfromtxt(
    "data/hr.csv",
    delimiter=",",
    names=True,
    dtype=None,
)

lum = arr["luminosity_class"]
lum_mapped = np.where(lum == "", "unknown", lum)
classes, counts = np.unique(lum_mapped, return_counts=True)

plt.bar(classes, counts)
plt.xlabel("Класс светимости")
plt.ylabel("Число звёзд")
plt.yscale("log")
plt.show()
```

### Диаграмма рассеяния на небесной сфере

<div class="alert alert-block alert-warning" style="margin-top: 20px">

<font size=4>**Задание 2**</font>

В файле `data/sga.csv` лежат данные обзора Siena Galaxy Atlas (те же, что были в прошлом ноутбуке).

Построить диаграмму, показывающую каждый объект на небесной сфере.

</div>

```python
import numpy as np
import matplotlib.pyplot as plt

arr = np.genfromtxt(
    "data/sga.csv",
    delimiter=",",
    names=True,
    dtype=None,
)

ra_rad = np.deg2rad(np.mod(arr["ra"] + 180, 360) - 180)
dec_rad = np.deg2rad(arr["dec"])

fig, ax = plt.subplots(subplot_kw={"projection": "aitoff"})
ax.plot(ra_rad, dec_rad, ",", linestyle="none", markersize=0.5, alpha=0.2)
ax.grid(True)
plt.show()
```

### Построение карт и `imshow`

<div class="alert alert-block alert-warning" style="margin-top: 20px">

<font size=4>**Задание 3**</font>

В файле `data/gaia.csv` лежат данные плотности звёзд в окрестности Солнца. Данные расположены в двухмерной диаграмме - `x_kpc` и `y_kpc` - это координаты соответствующих ячеек этой диаграммы. Колонка `overdensity` содержит нормированное значение плотности. 

Построить карту плотности в окрестности Солнца, отметить Солнце в координатах 0,0.

</div>

```python
import numpy as np
import matplotlib.pyplot as plt

d = np.genfromtxt("data/gaia.csv", delimiter=",", names=True)
x_u, x_inv = np.unique(d["x_kpc"], return_inverse=True)
y_u, y_inv = np.unique(d["y_kpc"], return_inverse=True)
Z = np.zeros((len(y_u), len(x_u)))
Z[y_inv, x_inv] = d["overdensity"]

half = 0.5 * (x_u[1] - x_u[0])
extent = (x_u[0] - half, x_u[-1] + half, y_u[0] - half, y_u[-1] + half)

fig, ax = plt.subplots()
im = ax.imshow(
    Z,
    extent=extent,
    origin="lower",
    aspect="equal",
    cmap="viridis",
)
ax.plot(0, 0, marker="*", color="yellow", markersize=14, markeredgecolor="black")
ax.set_xlabel("x, kpc")
ax.set_ylabel("y, kpc")
fig.colorbar(im, ax=ax, label="overdensity")
plt.show()
```
