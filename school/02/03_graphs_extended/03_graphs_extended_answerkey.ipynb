Это ещё один тип графиков. Его основная цель - сравнить между собой данные, приводимые к дискретному набору классов. Примеры таких задач:

#### Разница между гистограммой и столбчатой диаграммой

- Столбчатая диаграмма показывает, как распределено какое-то число в зависимости от *дискретного* значения. Дискретный, т.е. прерывистый - существует конечное число уникальных значений той величины, а зависимости от которой мы строим. В столбчатой диаграмме порядок значений на оси X чаще всего не важен.
- Гистограмма показывает распределение одной величины от *непрерывной* другой величины. Чтобы построить гистограмму, нужно перед построением разложить данные по "бакетам" (или "бинам") - дискретным диапазонам той величины, от которой мы строим зависимость. В гистограммах важно сохранить порядок величины, от которой идёт построение зависимости.

Для рисования столбчатых диаграмм в `matplotlib` есть отдельная функция (https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html):

````python
import matplotlib.pyplot as plt

classes = ["Ia", "Ibc", "II"]
percents = [25, 15, 60]

plt.bar(classes, percents)
plt.xlabel("Тип сверхновой")
plt.ylabel("Число событий, %")
plt.show()
````

Для гистограмм так же есть отдельная функция (https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html):

````python
import numpy as np
import matplotlib.pyplot as plt

# Гауссово (нормальное) распределение
data = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(data, bins=30) # bins - число бакетов/бинов, по которым разложить случайную величину
plt.xlabel("Значение случайной величины")
plt.ylabel("Число")
plt.show()
````

С гистограммами и столбчатыми диаграммами можно играться:

````python
import numpy as np
import matplotlib.pyplot as plt

# Случайная величина с экспоненциальным распределенем - вероятность выпадения величины x спадает по экспоненте
data = np.random.exponential(scale=10, size=1000)

plt.hist(data, 
    bins=30, 
    orientation='horizontal', #  горизонтальные столбцы
    edgecolor='red', #  цвет границ столбцов
    facecolor='white', #  цвет заливки столбцов
    rwidth=0.5, #  ширина каждого столбца - от 0 до 1, 1 - столбец занимает всё расстояние между двумя соседними столбцами
    linewidth=2, #  толщина границ столбцов
)
plt.xlabel("Число")
plt.ylabel("Значение случайной величины")
plt.show()
````

### Логарифмическая шкала

По графику выше видно, что если у значений большой разборс, отличить значения у хвостов распределения может быть сложно - на фоне столбца в 250 значений 2, 7 и 10 выглядят очень похоже, несмотря на то, что 10 в 5 раз больше 2.

В таких случаях помогает логарифмическая шкала - вместо того, чтобы по оси Y рисовать равноудалённые штрихи для 50, 100, 150, 200 (линейная шкала), давайте рисовать равноудалённые штрихи для величин, удалённых в 10 раз друг от друга (логарифмическая шкала): 10, 100, 1000, 10000. 

В `matplotlib` это делается одной командой:

```python
plt.yscale('log')
```

````python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.exponential(scale=10, size=1000)

plt.hist(data, 
    bins=30, 
    edgecolor='black',
    facecolor='white',
)

plt.xlabel("Значение случайной величины")
plt.ylabel("Число")
plt.yscale('log')
plt.show()
````

### Фильтрация данных

Помимо поэлементных математических операций `numpy` так же поддерживает и операции сравнения. Например:

````python
import numpy as np

arr = np.random.randint(100, 200, 20)

arr_filter = arr > 150
print(arr)
print(arr_filter)
````

В данном случае, `arr_filter` - это numpy-вектор булевых значений, отвечающий на вопрос "Какие элементы изначального вектора удовлетворяют условию `arr[i] > 150`?". Имея такой вектор, мы можем сделать новый numpy-вектор, в котором будут только элементы больше 150:

````python
new_arr = arr[arr_filter]

print(new_arr)
````

### Несколько графиков в одном

`matplotlib` позволяет рисовать несколько графиков в одном окне (на одной картинке) - полезно, если есть две связанные визуализации, которые хочется показывать одновременно. Чаще всего они должны разделять какую-то общую ось. Для возможности отдельной манипуляции каждым графиком есть функция `subplots`:

````python
import matplotlib.pyplot as plt

fig = plt.figure() # создать окно где будут графики

# разбить окно на сетку графиков - 2 ряда (первый аргумент), 
# 1 столбец (второй аргумент). В этой сетке добавить первый 
# график (третий аргумент).
ax1 = fig.add_subplot(2, 1, 1)

x = np.linspace(0, 10, 100)
ax1.plot(x, np.sin(x))

ax2 = fig.add_subplot(2, 1, 2)

ax2.plot(x, np.cos(x), 'r')

plt.show()
````

### Диаграмма рассеяния на небесной сфере

До сего момента мы строили всё на плоскости, определяемой осями X и Y. Не всегда можно натянуть все точки на такие координаты. Пример - небесная сфера. Для построения таких графиков в `matplotlib` есть специальная проекция - `aitoff` (использовать её можно только при помощи `subplots`):

````python
import numpy as np
import matplotlib.pyplot as plt

ra = np.linspace(0, 360, 200)
dec = 22 * np.sin(np.deg2rad(ra))

ra_aitoff = np.deg2rad(np.mod(ra + 180, 360) - 180)
dec_rad = np.deg2rad(dec)

fig = plt.figure()
# add_subplot без трёх численных аргументов сделает только один график
ax = fig.add_subplot(projection="aitoff")

ax.plot(ra_aitoff, dec_rad, label="Путь по небу", color='teal')
ax.grid(True)
ax.legend()
plt.show()
````

Построить диаграмму, показывающую объекты из каталога PGC на небесной сфере.

На дополнительном график построить гистограмму распределения числа PGC, NGC и SDSS объектов в каталоге и отдельным столбцом все остальные.

Примечание:
- Для `numpy`-вектора можно проверить, начинается ли он с определённой подстроки, при помощи функции `np.strings.startswith`:

    ```python
    arr = np.array(["abc", "bc", "ac"])
    arr_filter = np.strings.startswith(arr, "a")
    print(arr[arr_filter])
    ```

````python
import numpy as np
import matplotlib.pyplot as plt

arr = np.genfromtxt(
    "data/sga.csv",
    delimiter=",",
    names=True,
    dtype=None,
)

names = arr["name"]
is_pgc = np.strings.startswith(names, "PGC")
is_ngc = np.strings.startswith(names, "NGC")
is_sdss = np.strings.startswith(names, "SDSS")
n_pgc = int(is_pgc.sum())
n_ngc = int(is_ngc.sum())
n_sdss = int(is_sdss.sum())
is_other = len(arr) - n_pgc - n_ngc - n_sdss

ra_rad = np.deg2rad(np.mod(arr["ra"][is_pgc] + 180, 360) - 180)
dec_rad = np.deg2rad(arr["dec"][is_pgc])

fig = plt.figure()
ax_sky = fig.add_subplot(2, 1, 1, projection="aitoff")
ax_sky.plot(ra_rad, dec_rad, ",", linestyle="none", markersize=0.5, alpha=0.2)
ax_sky.grid(True)

ax_hist = fig.add_subplot(2, 1, 2)
labels = ["PGC", "NGC", "SDSS", "Остальные"]
counts = [n_pgc, n_ngc, n_sdss, is_other]
ax_hist.bar(labels, counts)
ax_hist.set_ylabel("Число объектов")

plt.tight_layout()
plt.show()
````

### Построение карт и `imshow`

<div class="alert alert-block alert-warning" style="margin-top: 20px">

<font size=4>**Задание 3**</font>

В файле `data/gaia.csv` лежат данные плотности звёзд в окрестности Солнца. Данные расположены в двухмерной диаграмме - `x_kpc` и `y_kpc` - это координаты соответствующих ячеек этой диаграммы. Колонка `overdensity` содержит нормированное значение плотности. 

Построить карту плотности в окрестности Солнца, отметить Солнце в координатах 0,0.

</div>

````python
import numpy as np
import matplotlib.pyplot as plt

d = np.genfromtxt("data/gaia.csv", delimiter=",", names=True)
x_u, x_inv = np.unique(d["x_kpc"], return_inverse=True)
y_u, y_inv = np.unique(d["y_kpc"], return_inverse=True)
Z = np.zeros((len(y_u), len(x_u)))
Z[y_inv, x_inv] = d["overdensity"]

half = 0.5 * (x_u[1] - x_u[0])
extent = (x_u[0] - half, x_u[-1] + half, y_u[0] - half, y_u[-1] + half)

fig, ax = plt.subplots()
im = ax.imshow(
    Z,
    extent=extent,
    origin="lower",
    aspect="equal",
    cmap="viridis",
)
ax.plot(0, 0, marker="*", color="yellow", markersize=14, markeredgecolor="black")
ax.set_xlabel("x, kpc")
ax.set_ylabel("y, kpc")
fig.colorbar(im, ax=ax, label="overdensity")
plt.show()
````
