# Переменные

Основная задача программирования - что-то посчитать. Чтобы это сделать, компьютеру надо объяснить две вещи:
- Откуда взять данные, над которыми провести расчёты
- Куда положить результат расчётов

Для этого и нужны переменные - это способ сообщить языку программирования, что у тебя в ячейке памяти лежит какое-то значение.

![](image.png)

Переменные специальным образом хранятся в оперативной памяти компьютера, но сейчас это не важно - можно туда о них как о коробках, в которые можно положить какое-то значение. Это значение может быть чем угодно - число, картинка, строка текста, видео или нечитаемый набор символов.

Для объявления переменных используется синтаксис `<имя переменной> = <значение переменной>`. Чтобы вывести переменную в ноутбуках, достаточно просто поставить её последней строкой. Например:

````python
a = 4

a
````

Переменные бывают нескольких основных типов:
- Целые числа
- Нецелые числа
- Строки (о них позже)
- Булевые значения (о них ещё позже)

# Строки

Помимо чисел, Python так же умеет работать со строками. Обычные операции над строками следующие:
- Сложение (умное слово - *конкатенация*). Каждая следующая складываемая строка складывается с предыдущей: `"я умею" + " говорить слова"`.
- Умножение на целое число. Строка будет повторена столько раз, на сколько она была умножена: `"я есть грут" * 5`.

<div class="alert alert-block alert-info" style="margin-top: 20px">

<font size=4>**Примечание**</font>     

Строка - это набор символов. Пробел - такой же один из возможных символов. При сложении и умножении строк никаких дополнительных символов не добавляется. Соответственно, никаких пробелов (если вы сами это не написали) добавлено не будет.

</div>

Например:

````python
first_words = "слова говорить"
last_words = "я умею"

concatenated_words = first_words + last_words
multiplied_words = first_words * 3 + last_words * 2

concatenated_words, multiplied_words
````

Если хочется пробелы, их надо добавлять самим:

````python
concatenated_words = first_words + " " + last_words
multiplied_words = (first_words + " ") * 3 + (last_words + " ") * 2

concatenated_words, multiplied_words
````

# Ввод и вывод

Вписанные напрямую в код (*захардкоженные*) значения - это, конечно, интересно, но хочется, чтобы поведение программы как-то менялось в зависимости от того, кто и зачем ей пользуется. Для этого в Python есть возможность ввода и вывода, которая позволяют пользователю ввести какие-то значения, а программисту - произвести над этими значениями какие-то действия (сложить, умножить, поделить и прочее) и вывести обратно результат.

Выводить мы уже умеем - достаточно в конце блока кода написать имя переменной (или нескольких переменных через запятую), и мы увидим значение этих переменных. Тем не менее, мы можем захотеть и посмотреть какие-то промежуточные значения, и в таком случае ничего не получится:

````python
moya_stroka = "вот она"
moya_vtoraya_stroka = "и тут она"

moya_stroka

moya_stroka_3 = moya_stroka + " " + moya_vtoraya_stroka

moya_stroka_3
````

Вывелась только `moya_stroka_3`, так как Jupyter notebook в целом выводит только то, что мы пишем в последней строке кода в блоке. Чтобы вывести значения из середины блока, есть специальная функция (aka код, написанный другим программистом) `print`, который позволяет вывести значения в середине блока:

````python
moya_stroka = "вот она"
moya_vtoraya_stroka = "и тут она"

print(moya_stroka)

moya_stroka_3 = moya_stroka + " " + moya_vtoraya_stroka

print(moya_stroka_3)
````

Это полезно, если хочется вывести какие-то промежуточные значения, а потом продолжать расчёты. Ещё один пример:

````python
I = 5 # ампер
U = 10 # вольт

R = U / I # Ом
print(R)

Q = I ** 2 * R # Дж
print(Q)

# Кстати, можно вывести сразу два значения, передав в print обе переменные (пробел подставится автоматически):
print(Q, R)
````

При помощи `print()` тексты можно печатать в несколько строк, для этого используются тройные двойные (хаха) кавычки:

````python
multiline_string = """Ну тут первая строка
А тут ну вторая
Смайлики кстати тоже можно💀💀"""

print(multiline_string)
````

Полноценно выводить научились, а что с получением ввода пользователя? Тут тоже есть специальная функция - `input()`. При выполнении написанного кода эта функция заставит код ждать ввода пользователя, а затем передаст значение, которое ввёл пользователь, в указанную переменную. В VSCode (где вы скорее всего сейчас) сверху появится поле для ввода, чтобы закончить ввод, надо нажать Enter. Пример:

````python
prefix = "Чего говоришь?"

sentence = input()

print(prefix + " " + sentence)
````

Часто надо получить какое-то число из ввода и произвести над ним математические действия. Просто так это сделать нельзя - нужно преобразовать его к численному типу при помощи функции `int()` или `float()`. Первая сделает из ввода целое число, вторая - число с точкой.

<details>
<summary>А почему нельзя?</summary>

Тут надо немного обсудить типы переменных в Python. Самых простых несколько:
- Строки
- Целые числа
- Числа с плавающей запятой
- Булевые значения

Функция `input()` всегда отдаёт строку. Да, эта строка может содержать одно число - "3". Тем не менее, Python не понимает, что мы хотим с этим дальше сделать.

Например, представим такой код:

```python
my_data = input()

print(my_data * 5)
```

Если во вводе пользователь введёт `100`, то без явного указания типа непонятно, что именно нужно вывести в выводе. Если `100` - это строка, то в выводе должно быть `100100100100100`, если же `100` - это число, то в выводе должно быть `500`. Python решает эту неоднозначность тем, что всегда интерпретирует ввод как строку, а если хочется это интерпретировать как число - это нужно явно указать при помощи `int()`. Таким образом, вывод кода выше будет `100100100100100`.

</details>

````python
my_string = input() # сюда введите своё имя

print("Привет,", my_string)

my_first_number = input() # сюда введите какое-то целое число

my_first_int = int(my_first_number)
print("Умножил на 3:", my_first_int * 3)

my_second_number = input() # сюда введите какое-то нецелое число (для разделения целой и дробной части используется именно точка, не запятая)

my_first_float = float(my_second_number)
print("Умножил на 5:", my_first_float * 5)
````
